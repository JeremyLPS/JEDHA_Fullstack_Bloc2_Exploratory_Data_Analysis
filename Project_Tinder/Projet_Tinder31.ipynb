# **Question 4 : "Can people accurately predict their own perceived value in the dating market?"**

```python
#cration du mask pour les colonnes correspondant au survey d'auto valuation
colonnes_2=[col for col in df_waves.columns if '3_1' in col]
self_perception = df_waves[colonnes_2]

#Calcul de la moyenne de chaque attribut aprs suppression des valeurs manquantes
self_perception.dropna().mean()
```

```python
#Colonnes des notes donnes par le partenaire
notes_partner = ['attr_o','sinc_o','intel_o','fun_o','amb_o']
rating_by_partner = df_waves[notes_partner]
rating_by_partner
```

```python
#Concatenation des deux df (perception personnelle + perception du partenaire) et supression des lignes avec valeurs manquantes
self_perception_2 = pd.concat([self_perception,rating_by_partner], axis=1)
self_perception_2 = self_perception_2.dropna()
self_perception_2.mean()
```

```python
#Rorganisation des index pour une meilleure visibilit des colonnes  comparer
self_perception_2.sort_index(axis=1, ascending=True)
self_perception_2[['diff_amb','diff_attr','diff_fun','diff_intel','diff_sinc']]=0
```

```python
#Calcul de la diffrence entre les notes donnes par soi-mme et par le partenaire pour chaque attribut
self_perception_2['diff_amb']= self_perception_2['amb_o']-self_perception_2['amb3_1']
self_perception_2['diff_attr'] = self_perception_2['attr_o']-self_perception_2['attr3_1']
self_perception_2['diff_fun'] = self_perception_2['fun_o']- self_perception_2['fun3_1']
self_perception_2['diff_intel'] = self_perception_2['intel_o']-self_perception_2['intel3_1']
self_perception_2['diff_sinc'] = self_perception_2['sinc_o']-self_perception_2['sinc3_1']
```

```python
self_perception_2.mean()
```

```python
#Diffrence moyenne entre la perception personnelle et la perception du partenaire
self_perception_2.iloc[:,-5:].abs().mean()
```

```python
delta_amb = (((self_perception_2['amb_o'])-(self_perception_2['amb3_1']))/(self_perception_2['amb3_1'])*100).abs().mean().round(2)
delta_amb
```

```python
delta_attr = (((self_perception_2['diff_attr']))/(self_perception_2['attr3_1'])*100).abs().mean().round(2)
delta_attr
```

```python
delta_fun = (((self_perception_2['diff_fun']))/(self_perception_2['fun3_1'])*100).abs().mean().round(2)
delta_fun
```

```python
delta_intel = (((self_perception_2['diff_intel'])/(self_perception_2['intel3_1']))*100).abs().mean().round(2)
delta_intel
```

```python
delta_sinc = (((self_perception_2['diff_sinc']))/(self_perception_2['sinc3_1'])*100).abs().mean().round(2)
delta_sinc
```

```python
diff_notes = pd.DataFrame({"Attractive" : [self_perception_2['attr3_1'].mean(),self_perception_2['attr_o'].mean()],
              "Ambitious" : [self_perception_2['amb3_1'].mean(),self_perception_2['amb_o'].mean()],
              "Sincere" : [self_perception_2['sinc3_1'].mean(),self_perception_2['sinc_o'].mean()],
              "Fun" : [self_perception_2['fun3_1'].mean(),self_perception_2['fun_o'].mean()],
              "Intelligent" : [self_perception_2['intel3_1'].mean(),self_perception_2['intel_o'].mean()]
             },
                index =['Moyenne note auto-attribue','Moyenne note donne par partenaire']
            ).T
diff_notes
```

```python
#Graphe de la diffrence de note moyenne pour chaque attributs entre perception personnelle et perception du partenaire rencontr
sns.lineplot(diff_notes, markers=True, markersize=10,palette=['navy','brown'])
plt.xlabel('Attributs',fontweight='bold')
plt.ylabel('Moyenne',fontweight='bold')
plt.title('Comparaison entre notes moyennes auto-attribues et \n donnes par le partenaire rencontr',fontweight='bold')

plt.legend(bbox_to_anchor=(1, 0.5, 0.5, 0.5))
plt.show();
```

### ==> En moyenne, les participants ont tendance  lgrement se surrvaluer par rapport  ce que pense le partenaire rencontr et ce pour chaque attribut
### De manire globale, les participants s'attribuent une note pour chaque attribut 1 point plus haut que ce que leur partenaire a lui not.
### L'attribut "Fun" est la caractristique la plus surrvalue par les participants, l'ambition tant la moins surrvalue.

```python
# Rcupration des % de matches pour chaque dernier date pour chaque vague

liste = [w1,w2,w3,w4,w5]
j=1
for i in liste :
    print("Pour la VAGUE {}, le % de match au premier et au dernier rdv est (respectivement) de :".format(j))
    print(round((i.groupby('order').get_group(i['order'].min())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    print()
    print(round((i.groupby('order').get_group(i['order'].max())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    j=j+1
    print();

liste2 = [w10,w11,w12,w13,w14,w15,w16,w17,w18,w19,w20,w21]
a = 10
for k in liste2 :
    print("Pour la VAGUE {}, le % de match au premier et au dernier rdv est (respectivement) de :".format(a))
    print(round((k.groupby('order').get_group(k['order'].min())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    print()
    print(round((k.groupby('order').get_group(k['order'].max())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    a+=1
    print();
```

### Sur les 17 vagues retenues, 7 ont dbouch sur un % de match plus lev  la premiere qu' la dernire rencontre de la soire.
###  l'inverse, 7 autres vagues ont dbouch sur un % de match plus lev  la dernire qu' la premire rencontre de la soire.
### Enfin, 3 vagues ont dbouch sur un % quivalent entre les deux RDV (1er et dernier)

### ==> Ainsi, il n'y a pas de tendance claire qui se dgage attestant qu'une position dans l'ordre des rencontres est plus favorable  un match qu'une autre

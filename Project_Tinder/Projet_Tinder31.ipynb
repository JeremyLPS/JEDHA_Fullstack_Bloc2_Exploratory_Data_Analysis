# **Question 2 : "How important do people think attractiveness is in potential mate selection vs. its real impact?"**

```python
#Note moyenne donnée par les femmes pour l'attribut Attractiveness chez un homme
note_female_before = df_waves[female]['attr1_1'].mean().round(2)
note_female_before
```

```python
#Note moyenne donnée par les femmes pour l'attribut Attractiveness chez un homme après le speed dating
note_female_after = df_waves[female]['attr7_2'].mean().round(2)
note_female_after
```

```python
#Note moyenne donnée par les hommes pour l'attribut Attractiveness chez une femme avant le speed dating
note_male_before = df_waves[male]['attr1_1'].mean().round(2)
note_male_before
```

```python
#Note moyenne donnée par les hommes pour l'attribut Attractiveness chez une femme après le speed dating
note_male_after = df_waves[male]['attr7_2'].mean().round(2)
note_male_after
```

```python
#Delta entre note avant et note après speed dating
delta_femmes = round((note_female_after - note_female_before)/note_female_before*100,2)
delta_femmes
```

```python
#Delta entre note avant et note après speed dating
delta_hommes = round((note_male_after - note_male_before)/note_male_before*100,2)
delta_hommes
```

```python
print("La note moyenne des femmes avant le speed dating concernant l'importance de l'attribut Attractiveness \n"
      "chez un homme était de :", note_female_before)
print("Après le speed dating cette note monte à :", note_female_after, "soit un impact réel", delta_femmes,"% supérieur")

print()

print("La note moyenne des hommes avant le speed dating concernant l'importance de l'attribut Attractiveness \n"
      "chez une femme était de :", note_male_before)
print("Après le speed dating cette note monte à :", note_male_after, "soit un impact réel", delta_hommes,"% supérieur")
```

```python
notes = pd.DataFrame({'Hommes' : [note_male_before,note_male_after],
                      'Femmes' : [note_female_before,note_female_after]
                     },
                     index =['Notes avant speed dating','Notes après speed dating']

                    )
notes
```

# **Question 4 : "Can people accurately predict their own perceived value in the dating market?"**

```python
#création du mask pour les colonnes correspondant au survey d'auto évaluation
colonnes_2=[col for col in df_waves.columns if '3_1' in col]
self_perception = df_waves[colonnes_2]

#Calcul de la moyenne de chaque attribut après suppression des valeurs manquantes
self_perception.dropna().mean()
```

```python
#Colonnes des notes données par le partenaire
notes_partner = ['attr_o','sinc_o','intel_o','fun_o','amb_o']
rating_by_partner = df_waves[notes_partner]
rating_by_partner
```

```python
#Concatenation des deux df (perception personnelle + perception du partenaire) et supression des lignes avec valeurs manquantes
self_perception_2 = pd.concat([self_perception,rating_by_partner], axis=1)
self_perception_2 = self_perception_2.dropna()
self_perception_2.mean()
```

```python
#Réorganisation des index pour une meilleure visibilité des colonnes à comparer
self_perception_2.sort_index(axis=1, ascending=True)
self_perception_2[['diff_amb','diff_attr','diff_fun','diff_intel','diff_sinc']]=0
```

```python
#Calcul de la différence entre les notes données par soi-même et par le partenaire pour chaque attribut
self_perception_2['diff_amb']= self_perception_2['amb_o']-self_perception_2['amb3_1']
self_perception_2['diff_attr'] = self_perception_2['attr_o']-self_perception_2['attr3_1']
self_perception_2['diff_fun'] = self_perception_2['fun_o']- self_perception_2['fun3_1']
self_perception_2['diff_intel'] = self_perception_2['intel_o']-self_perception_2['intel3_1']
self_perception_2['diff_sinc'] = self_perception_2['sinc_o']-self_perception_2['sinc3_1']
```

```python
self_perception_2.mean()
```

```python
#Différence moyenne entre la perception personnelle et la perception du partenaire
self_perception_2.iloc[:,-5:].abs().mean()
```

```python
delta_amb = (((self_perception_2['amb_o'])-(self_perception_2['amb3_1']))/(self_perception_2['amb3_1'])*100).abs().mean().round(2)
delta_amb
```

```python
delta_attr = (((self_perception_2['diff_attr']))/(self_perception_2['attr3_1'])*100).abs().mean().round(2)
delta_attr
```

```python
delta_fun = (((self_perception_2['diff_fun']))/(self_perception_2['fun3_1'])*100).abs().mean().round(2)
delta_fun
```

```python
delta_intel = (((self_perception_2['diff_intel'])/(self_perception_2['intel3_1']))*100).abs().mean().round(2)
delta_intel
```

```python
delta_sinc = (((self_perception_2['diff_sinc']))/(self_perception_2['sinc3_1'])*100).abs().mean().round(2)
delta_sinc
```

```python
diff_notes = pd.DataFrame({"Attractive" : [self_perception_2['attr3_1'].mean(),self_perception_2['attr_o'].mean()],
              "Ambitious" : [self_perception_2['amb3_1'].mean(),self_perception_2['amb_o'].mean()],
              "Sincere" : [self_perception_2['sinc3_1'].mean(),self_perception_2['sinc_o'].mean()],
              "Fun" : [self_perception_2['fun3_1'].mean(),self_perception_2['fun_o'].mean()],
              "Intelligent" : [self_perception_2['intel3_1'].mean(),self_perception_2['intel_o'].mean()]
             },
                index =['Moyenne note auto-attribuée','Moyenne note donnée par partenaire']
            ).T
diff_notes
```

```python
#Graphe de la différence de note moyenne pour chaque attributs entre perception personnelle et perception du partenaire rencontré
sns.lineplot(diff_notes, markers=True, markersize=10,palette=['navy','brown'])
plt.xlabel('Attributs',fontweight='bold')
plt.ylabel('Moyenne',fontweight='bold')
plt.title('Comparaison entre notes moyennes auto-attribuées et \n données par le partenaire rencontré',fontweight='bold')

plt.legend(bbox_to_anchor=(1, 0.5, 0.5, 0.5))
plt.show();
```

### ==> En moyenne, les participants ont tendance à légèrement se surrévaluer par rapport à ce que pense le partenaire rencontré et ce pour chaque attribut
### De manière globale, les participants s'attribuent une note pour chaque attribut 1 point plus haut que ce que leur partenaire a lui noté.
### L'attribut "Fun" est la caractéristique la plus surrévaluée par les participants, l'ambition étant la moins surrévaluée.

# **Question 5 : "In terms of getting a second date, is it better to be someone's first speed date of the night or their last?"**

```python
#Pourcentage de match par vague et par position du date dans la soirée
round((df_waves.groupby(['wave','order'])['match'].value_counts(normalize=True)*100),2)
```

```python
round(pd.crosstab(df_waves['order'],df_waves['match'], normalize='index')*100,2)
```

```python
# Récupération des % de matches pour chaque dernier date pour chaque vague

liste = [w1,w2,w3,w4,w5]
j=1
for i in liste :
    print("Pour la VAGUE {}, le % de match au premier et au dernier rdv est (respectivement) de :".format(j))
    print(round((i.groupby('order').get_group(i['order'].min())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    print()
    print(round((i.groupby('order').get_group(i['order'].max())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    j=j+1
    print();

liste2 = [w10,w11,w12,w13,w14,w15,w16,w17,w18,w19,w20,w21]
a = 10
for k in liste2 :
    print("Pour la VAGUE {}, le % de match au premier et au dernier rdv est (respectivement) de :".format(a))
    print(round((k.groupby('order').get_group(k['order'].min())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    print()
    print(round((k.groupby('order').get_group(k['order'].max())['match']
                 .value_counts(normalize=True)*100),2).to_string()) #to_string pour enlever l'affichage de name
                                                                    # et dtype dans les prints
    a+=1
    print();
```

### Sur les 17 vagues retenues, 7 ont débouché sur un % de match plus élevé à la premiere qu'à la dernière rencontre de la soirée.
### À l'inverse, 7 autres vagues ont débouché sur un % de match plus élevé à la dernière qu'à la première rencontre de la soirée.
### Enfin, 3 vagues ont débouché sur un % équivalent entre les deux RDV (1er et dernier)

### ==> Ainsi, il n'y a pas de tendance claire qui se dégage attestant qu'une position dans l'ordre des rencontres est plus favorable à un match qu'une autre
